# Transformation

```python
transactions_data['t_dat'] = pd.to_datetime(transactions_data.t_dat)
```

```python
transactions_data_agg = transactions_data.groupby(
    [
        'customer_id',
        'article_id'
    ],as_index=False
).agg(
    purchase_count=('price','count'),
    purchase_amount=('price',sum)
)

transactions_data_agg['client_id'] = pd.factorize(transactions_data_agg['customer_id'])[0]

transactions_data_agg.head()
```

```python
ids_data = transactions_data_agg[['customer_id','client_id']].drop_duplicates()

ids_data.head()
```

## Agg por fecha

```python
agg_article_date = transactions_data.groupby(
    [
        pd.Grouper(
            key='t_dat',
            freq='1W'
        ),
        'article_id'
    ],
    # as_index=False
).agg(
    purchase_count=('price','count'),
    purchase_amount=('price',sum)
).reset_index()

agg_article_date = pd.merge(
    agg_article_date,
    articles_data[
        [
            'article_id',
            'product_group_name'#'product_type_name'
            
        ]
    ],
    how='left',
    on='article_id'
).rename(
    columns = {
        'product_group_name':'product_type_name'
    }
)

agg_article_date = agg_article_date.groupby(
    [
        pd.Grouper(
            key='t_dat',
            freq='1W'
        ),
        'product_type_name'
    ],
    # as_index=False
).agg(
    purchase_count=('purchase_count','sum'),
    purchase_amount=('purchase_amount',sum)
).reset_index()

agg_article_date['avg_purchase'] = agg_article_date['purchase_amount']/agg_article_date['purchase_count']



agg_article_date.head()
```

```python
agg_article = agg_article_date.groupby('product_type_name').agg(
    purchase_count=('purchase_count','sum'),
    purchase_amount=('purchase_amount',sum)
).reset_index()

agg_article['avg_purchase'] = agg_article_date['purchase_amount']/agg_article_date['purchase_count']

agg_article.head()
```

```python
agg_date = agg_article_date.groupby('t_dat').agg(
    purchase_count=('purchase_count','sum'),
    purchase_amount=('purchase_amount',sum)
).reset_index()

agg_date['avg_purchase'] = agg_date['purchase_amount']/agg_date['purchase_count']

agg_date.head()
```

```python
agg_weekdate_article = transactions_data.groupby(
    [
        transactions_data.t_dat.dt.day_name(),
        'article_id'
    ]
).agg(
    purchase_count=('price','count'),
    purchase_amount=('price',sum)
).reset_index()

agg_weekdate_article['avg_purchase'] = agg_weekdate_article['purchase_amount']/agg_weekdate_article['purchase_count']


agg_weekdate_article = pd.merge(
    agg_weekdate_article,
    articles_data[
        [
            'article_id',
            'product_group_name'#'product_type_name'
            
        ]
    ],
    how='left',
    on='article_id'
).rename(
    columns = {
        'product_group_name':'product_type_name'
    }
)

agg_weekdate_article = agg_weekdate_article.groupby(
    [
        't_dat',
        'product_type_name'
    ],
    # as_index=False
).agg(
    purchase_count=('purchase_count','sum'),
    purchase_amount=('purchase_amount',sum)
).reset_index()

agg_weekdate_article['avg_purchase'] = agg_weekdate_article['purchase_amount']/agg_weekdate_article['purchase_count']

agg_weekdate_article.head()
```
